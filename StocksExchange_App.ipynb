```python
import pandas as pd
def load_data():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header = 0)
    data = html[0]
    return data
```

```python
data = load_data()
data.head(5)
```

```python
import yfinance as yf
```

```python
new_data = yf.download(
    tickers = list(data.Symbol),
    period = 'ytd',
    interval = '1d',
    group_by = 'ticker',
    auto_adjust= True,
    prepost= True)
```

```python
new_data
```

```python
new_data['AAPL']
```

```python
df = pd.DataFrame(new_data['AAPL'].Close)
df['Date'] = df.index
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df.Date, df.Close, color='skyblue', alpha=0.8)
plt.xticks(rotation = 90)
plt.xlabel('Date')
plt.ylabel('Close')
```

```python
def price_plot(Symbol):
    df = pd.DataFrame(new_data[Symbol].Close)
    df['Date'] = df.index
    plt.plot(df.Date, df.Close, color='skyblue', alpha=0.8)
    plt.xticks(rotation = 90)
    plt.title(Symbol, fontweight='bold')
    plt.xlabel('Date')
    plt.ylabel('Close')
    return plt.show()
```

```python
price_plot('AMGN')
```
